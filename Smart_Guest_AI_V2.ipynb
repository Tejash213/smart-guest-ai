```python
import gradio as gr
from deepface import DeepFace
from tensorflow.keras.models import load_model
from PIL import Image, ImageOps
import numpy as np

# Load model and labels once
model = load_model("/content/smart_guest_ai_v2/models/keras_model.h5")
with open("/content/smart_guest_ai_v2/models/labels.txt", "r") as f:
    labels = [line.strip() for line in f.readlines()]

# Suggest music
def suggest_music(emotion, dressing):
    if emotion == "happy":
        if dressing == "punjabi":
            return "üé∂ Play Bhangra party mix"
        elif dressing == "formal":
            return "üé∂ Play instrumental jazz"
        else:
            return "üé∂ Play happy pop hits"
    elif emotion == "sad":
        return "üé∂ Play soothing soft music"
    elif emotion == "angry":
        return "üé∂ Play calming acoustic songs"
    elif emotion == "surprise":
        return "üé∂ Play upbeat electronic vibes"
    else:
        return "üé∂ Play chill background lo-fi"

# AI prediction function
def analyze_guest(image):
    # Emotion
    emotion_result = DeepFace.analyze(img_path=image, actions=['emotion'], enforce_detection=False)
    emotion = emotion_result[0]['dominant_emotion']

    # Dressing style
    img = Image.open(image).convert("RGB")
    img = ImageOps.fit(img, (224, 224), Image.Resampling.LANCZOS)
    img_array = np.asarray(img) / 255.0
    img_array = np.expand_dims(img_array, axis=0)
    predictions = model.predict(img_array)
    predicted_index = np.argmax(predictions)
    dressing = labels[predicted_index].split(" ", 1)[-1]

    music = suggest_music(emotion, dressing)

    return f"üß† Emotion: {emotion}\nüß• Style: {dressing}\nüé∂ Music: {music}"

# Create interface
gr.Interface(
    fn=analyze_guest,
    inputs=gr.Image(type="filepath"),
    outputs="text",
    title="Smart Guest AI",
    description="Upload a guest image to get emotion, dressing style, and music suggestion!"
).launch()
```
